```python
import numpy as np
import pandas as pd
pd.set_option('display.max_columns', None)
import matplotlib.pylab as plt
%matplotlib inline
```

```python
### LGB - 3
lgb1 = pd.read_csv('sub_lgb.csv.gz')
lgb2 = pd.read_csv('sub_lgb2.csv.gz')
lgb3 = pd.read_csv('sub_lgb3.csv.gz')
lgb4 = pd.read_csv('sub_lgb4_grid.csv.gz')

lgb1.rename(columns={"meter_reading": "lgb1"}, inplace=True)
lgb2.rename(columns={"meter_reading": "lgb2"}, inplace=True)
lgb3.rename(columns={"meter_reading": "lgb3"}, inplace=True)
lgb4.rename(columns={"meter_reading": "lgb4"}, inplace=True)
```

```python
sub = lgb1.merge(lgb2, on="row_id", how="left")
sub = sub.merge(lgb3, on="row_id", how="left")
sub = sub.merge(lgb4, on="row_id", how="left")
sub.head()
```

```python
### XGB - 3
xgb1 = pd.read_csv('sub_xgb.csv.gz')
xgb2 = pd.read_csv('sub_xgb2.csv.gz')
xgb3 = pd.read_csv('sub_xgb3.csv.gz')

xgb1.rename(columns={"meter_reading": "xgb1"}, inplace=True)
xgb2.rename(columns={"meter_reading": "xgb2"}, inplace=True)
xgb3.rename(columns={"meter_reading": "xgb3"}, inplace=True)
```

```python
sub = sub.merge(xgb1, on="row_id", how="left")
sub = sub.merge(xgb2, on="row_id", how="left")
sub = sub.merge(xgb3, on="row_id", how="left")
sub.head()
```

```python
### CAT - 3
cat1 = pd.read_csv('sub_cat.csv.gz')
cat2 = pd.read_csv('sub_cat2.csv.gz')
cat3 = pd.read_csv('sub_cat3.csv.gz')

cat1.rename(columns={"meter_reading": "cat1"}, inplace=True)
cat2.rename(columns={"meter_reading": "cat2"}, inplace=True)
cat3.rename(columns={"meter_reading": "cat3"}, inplace=True)
```

```python
sub = sub.merge(cat1, on="row_id", how="left")
sub = sub.merge(cat2, on="row_id", how="left")
sub = sub.merge(cat3, on="row_id", how="left")
sub.head()
```

```python
### BUILDING - 3
building1 = pd.read_csv('building_wise.csv.gz')
building2 = pd.read_csv('building_wise2.csv.gz')
building3 = pd.read_csv('building_wise3.csv.gz')

building1.rename(columns={"meter_reading": "building1"}, inplace=True)
building2.rename(columns={"meter_reading": "building2"}, inplace=True)
building3.rename(columns={"meter_reading": "building3"}, inplace=True)
```

```python
sub = sub.merge(building1, on="row_id", how="left")
sub = sub.merge(building2, on="row_id", how="left")
sub = sub.merge(building3, on="row_id", how="left")
sub.head()
```

```python
sub['meter_reading'] = sub.lgb1 * 0.07 + sub.lgb2 * 0.07 + sub.lgb3 * 0.07 + sub.lgb4 * 0.07 + \
                        sub.xgb1 * 0.07 + sub.xgb2 * 0.07 + sub.xgb3 * 0.07 + \
                        sub.cat1 * 0.07 + sub.cat2 * 0.07 + sub.cat3 * 0.07 + \
                        sub.building1 * 0.1 + sub.building2 * 0.1 + sub.building3 * 0.1
```

```python
sub.head()
```

```python
submit = sub[["row_id", "meter_reading"]]
submit.meter_reading = submit.meter_reading.round(4)
print(submit.shape)
submit.head()
```
